`NumPy` は配列を操作するための Python ライブラリです。また、線形代数、フーリエ変換、行列の領域で作業するための関数も備えています。 `NumPy` は2005年にTravis Oliphantによって作成されたオープンソースのプロジェクトであり、自由に使用することができます。 `NumPy` は Numerical Python の略です。

NumPyを使えば簡単に配列を作ることができます:

`Matplotlib` is an open-source drawing library that supports various drawing types. You can generate plots, histograms, bar charts, and other types of charts with just a few lines of code.

`Matplotlib` は、様々な描画タイプをサポートするオープンソースの描画ライブラリです。数行のコードでプロット、ヒストグラム、棒グラフなどを生成することができます。

An empty plot.

## Learn by the example: The Monty Hall problem

*Suppose you're on a game show, and you're given the choice of three doors: Behind one door is a car; behind the others, goats. You pick a door, say No. 1, and the host, who knows what's behind the doors, opens another door, say No. 3, which has a goat. He then says to you, "Do you want to pick door No. 2?" Is it to your advantage to switch your choice?* (source: Wikipedia)

*ゲーム番組で、3つのドアを選ぶように言われたとします。1つのドアの向こうには車、もう1つのドアの向こうにはヤギ。あなたは1番のドアを選び、ドアの向こうに何があるかを知っている司会者は、別のドア、例えば3番のドアを開け、そこにはヤギがいるとします。そして、「2番のドアを選びますか？」とあなたに言います。選択肢を変えることは、あなたにとって有利なことでしょうか？* (source: Wikipedia)



This case is a very famous problem involving probabilities that is counter-intuitive (except if you are comfortable with probabilities). 

このケースは、（確率に慣れている人は別として）直感に反する確率を含む非常に有名な問題です。

```python
from random import randint, seed

# Enum is a class in python for creating enumerations, which are a set of symbolic names (members) bound to unique,
# constant values.
from enum import Enum

# We create a class of Enum, which contains the two different strategies
class Strategy(Enum):
    CHANGE = 1
    KEEP = 2
```

The code above implements what we will need for our code.

上のコードは、私たちのコードに必要なものを実装したものです。

```python
seed()

def play_game(strategy):

    doors = [0, 1, 2]
    
    good_door = randint(0,2)
    
    # Player's choice
    first_choice = randint(0,2)
    
    # there are two doors left
    doors.remove(first_choice)
    
    # The host eliminates a door
    if first_choice == good_door:
        doors.remove(doors[randint(0,1)])
    else:
        doors = [good_door]
    
    second_choice = 0
    # Second choice depends on the strategy
    if strategy == Strategy.CHANGE:
        second_choice = doors[0]
    elif strategy == Strategy.KEEP:
        second_choice = first_choice
    else:
        raise ValueError("Unknown strategy!")
    
    return second_choice == good_door

def play(strategy, nb):
    return [1 if play_game(strategy) else 0 for i in range(nb)]
```

The code above implements the first part of our code: we generate one game. Now we need to generate a lot of games to see what is the best choice between `KEEP` and `CHANGE`.

上のコードは最初の部分を実装したもので、1つのゲームを生成しています。今度は、`KEEP`と`CHANGE`のどちらが良い選択なのかを調べるために、たくさんのゲームを生成する必要があります。

```python
print("By changing his choice, the player has won {} over 10000 parties."
      .format(sum(play(Strategy.CHANGE, 10000))))
      
print("By keeping his choice, the player has won {} over 10000 parties."
      .format(sum(play(Strategy.KEEP, 10000))))
```

#### Conclusion / 結論:

We can see that that by changing his choice, the player wins around 2/3 of the games while by keeping the same choice, the player wins only 1/3 of the games. We can conclude that **the right strategy is too change your choice!**.

Let's visualize it with `Matplotlib`.

選択肢を変えることによって勝率は2/3程度となり、同じ選択肢のままでは1/3程度の勝率になることがわかります。結論として、**正しい戦略は選択を変えることである！**と言えます。

これを `Matplotlib` で可視化してみよう。

```python
plot = plt.bar([1,2],[sum(play(Strategy.CHANGE, 10000)), 
               sum(play(Strategy.KEEP, 10000))], 
        tick_label=["Change","Keep"])
```

```python
profit_change = []
profit_keep = []
samples =  [1000, 10000, 20000, 50000, 80000, 100000]
for steps in samples:
    profit_change.append(play(Strategy.CHANGE, steps))
    profit_keep.append(play(Strategy.KEEP, steps))


plot = plt.scatter(samples, [sum(x) for x in profit_change], color='orange')
plot = plt.scatter(samples, [sum(x) for x in profit_keep])
plt.axis([0,105000,0, 80000]);
```
