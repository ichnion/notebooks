## Seaborn

`Seaborn` is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. It is also really useful because it works with dataframes from `pandas` library.

`Seaborn` は matplotlib をベースにした Python のデータ可視化ライブラリです。魅力的で情報量の多い統計グラフィックを描くための高レベルなインタフェースを提供します。また、 `pandas` ライブラリのデータフレームを扱うことができるので、非常に便利です。

```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

We use `load_dataset()` to load an example of a dataframe. Usually libraries have different dataframes usable to train ourselves with Machine Learning. Here we use `tips` dataset.

データフレームの例を読み込むには `load_dataset()` を使用します。通常、ライブラリには機械学習の学習に利用できる様々なデータフレームがあります。ここでは、`tips`データセットを使用します。



This plot shows the relationship between five variables in the tips dataset using a single call to the seaborn function `relplot()`. Notice how we provided only the names of the variables and their roles in the plot. Unlike when using matplotlib directly, it wasn’t necessary to specify attributes of the plot elements in terms of the color values or marker codes. Behind the scenes, seaborn handled the translation from values in the dataframe to arguments that matplotlib understands. This declarative approach lets you stay focused on the questions that you want to answer, rather than on the details of how to control matplotlib.

このプロットは、seaborn関数 `relplot()` を一回呼び出すだけで、Tipsデータセットの5つの変数の関係を表示します。変数の名前とプロットでの役割だけを提供したことに注意してください。matplotlibを直接使うときとは異なり、プロット要素の属性を色値やマーカーコードで指定する必要はありません。裏側では、seabornがデータフレーム内の値からmatplotlibが理解できる引数への変換を処理しました。この宣言的なアプローチにより、matplotlib の制御方法の詳細ではなく、答えたい質問に集中することができます。

Often, we are interested in the average value of one variable as a function of other variables. Many seaborn functions will automatically perform the statistical estimation that is necessary to answer these questions:

しばしば、ある変数の平均値を他の変数の関数として求めることがある。多くのseaborn関数は、このような疑問に答えるために必要な統計的推定を自動的に実行します:

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
```

#### Composite views onto multivariate datasets / 多変量データセットへの合成表示
Some `seaborn` functions combine multiple kinds of plots to quickly give informative summaries of a dataset. One, `jointplot()`, focuses on a single relationship. It plots the joint distribution between two variables along with each variable’s marginal distribution:

いくつかの `seaborn` 関数は、複数の種類のプロットを組み合わせて、データセットの情報を素早く要約することが可能です。そのうちの1つである `jointplot()` は、1つの関係に着目したものです。これは、2つの変数の間の結合分布と、それぞれの変数の周辺分布を描画します。

```python
penguins = sns.load_dataset("penguins")
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

We use the `penguins` dataframe:


The other, `pairplot()`, takes a broader view: it shows joint and marginal distributions for all pairwise relationships and for each variable, respectively:

もう一つの `pairplot()` はより広い視野を持ち、すべてのペアワイズ関係と各変数のジョイント分布とマージナル分布をそれぞれ表示します。

```python
sns.pairplot(data=penguins, hue="species")
```

For more examples, [check the Seaborn documentation](https://seaborn.pydata.org/examples/index.html).

## Pandas

`Pandas` is an open-source library that is made mainly for working with relational or labeled data both easily and intuitively. It provides various data structures and operations for manipulating numerical data and time series. This library is built on top of the `NumPy` library. `Pandas` is fast and it has high performance & productivity for users.

`Pandas`はリレーショナルデータやラベル付けされたデータを簡単かつ直感的に操作するために作られたオープンソースのライブラリです。数値データや時系列データを操作するための様々なデータ構造と操作を提供します。このライブラリは `NumPy` ライブラリの上に構築されています。Pandasは高速で、高いパフォーマンスと生産性を提供します。

The `Pandas` package is the most important tool at the disposal of Data Scientists and Analysts working in Python today. The powerful machine learning and glamorous visualization tools may get all the attention, but `Pandas` is the backbone of most data projects.

`Pandas`パッケージは、今日、Pythonで作業するデータサイエンティストやアナリストが自由に使える最も重要なツールです。強力な機械学習や華やかな可視化ツールが注目されるかもしれませんが、`Pandas`はほとんどのデータプロジェクトのバックボーンとなっています。
